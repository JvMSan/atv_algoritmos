```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('resultados_erros.csv')
df
```

```python
tipos = df['tipo'].unique()
for tipo in tipos:
    dados = df[df['tipo']==tipo]
    sns.set(style="white")
    # plt.rcParams["figure.figsize"] = (10, 6)
    # plt.figure()
    plt.plot(dados["n"], dados["erro_real"], label="Erro Relativo Real", marker='o')
    plt.plot(dados["n"], dados["erro_estimado"], label="Erro Relativo Estimado", marker='x')
    plt.xscale("log")
    plt.yscale("log")
    plt.title(f"Erro Relativo - Tipo: {tipo}")
    plt.xlabel("n (log)")
    plt.ylabel("Erro Relativo (log)")
    plt.legend()
    plt.tight_layout()
    plt.grid(True, which="both", ls="--")
    plt.savefig(f"erro_{tipo}.png")
    plt.show()
```
